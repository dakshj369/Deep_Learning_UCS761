```python
T=1.0
F=0.0

def get_or_data():
  X=[
      [F,F],
      [F,T],
      [T,F],
      [T,T]
  ]
  Y=[
      [F],
      [T],
      [T],
      [T]
  ]

  return X,Y


def get_and_data():
  X=[
      [F,F],
      [F,T],
      [T,F],
      [T,T]
  ]
  Y=[
      [F],
      [F],
      [F],
      [T]
  ]

  return X,Y


def get_xor_data():
  X=[
      [F,F],
      [F,T],
      [T,F],
      [T,T]
  ]
  Y=[
      [F],
      [T],
      [T],
      [F]
  ]

  return X,Y


```

```python
class Myprceptron:
  def __init__(self, eta=0.1, epochs=1000):
    self.weights =None
    self.bias = None
    self.eta = eta
    self.epochs = epochs

  def fit(self, X, y):
    self.weights = np.zeros(X.shape[1])
    self.bias = 0

    for epoch in range(self.epochs):
      for i in range(X.shape[0]):
        y_pred = self.Step_activ_func(np.dot(self.weights, X[i]) + self.bias)
        mae=Y[i] - y_pred
        self.weights += self.eta * (mae) * X[i]
        self.bias += self.eta * mae

  def Step_activ_func(self, x):
    if x > 0:
      return 1
    else:
      return 0

  def predict(self, X):
    predictions = []
    for i in range(X.shape[0]):
      prediction = self.Step_activ_func(np.dot(self.weights, X[i]) + self.bias)
      predictions.append(prediction)
    return np.array(predictions).reshape(-1, 1)
```

```python

clf=Myprceptron()
X,Y=get_or_data()
X=np.array(X)
Y=np.array(Y)
clf.fit(X,Y)

x_test=[
    [F,F],
    [T,F],
    [F,T],
    [F,F]
]
y_test=[
    [F],
    [T],
    [T],
    [F]
]

X_test=np.array(x_test)
Y_test=np.array(y_test)

Ypred=clf.predict(X_test)
print(Ypred)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import tensorflow as tf
import numpy as np

```

```python
T=1.0
F=0.0

def get_xor_data():
  X=[
      [F,F],
      [F,T],
      [T,F],
      [T,T]
  ]
  Y=[
      [F],
      [T],
      [T],
      [F]
  ]

  return X,Y
X,Y=get_xor_data()


x_test=[
    [T,F],
    [T,T],
    [F,T],
    [F,F]
]
y_test=[
    [T],
    [F],
    [T],
    [F]
]

# Convert data to numpy arrays
X = np.array(X)
Y = np.array(Y)
x_test = np.array(x_test)
y_test = np.array(y_test)
```

```python
model=Sequential()
model.add(Dense(16, input_dim=2, activation='relu'))
model.add(Dense(16, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
model.compile(loss='binary_crossentropy', optimizer='SGD', metrics=['accuracy'])
```

```python
model.fit(X,Y, epochs=1000)
```

```python
loss,accc=model.evaluate(x_test,y_test,verbose=0)
print(accc)
print(loss)
```
